```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)


print(df.head())

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
plt.figure(figsize=(12, 8))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(x=df[feature], y=[0]*df.shape[0], hue=df['species'])
    plt.title(f'Univariate Analysis: {feature}')
    plt.xlabel(feature)
    plt.ylabel('Constant (0)')
    plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
pairs = [
    ('sepal_width', 'petal_width'),
    ('sepal_length', 'petal_width'),
    ('sepal_length', 'petal_length')
]
plt.figure(figsize=(15, 5))
for i, (x_feat, y_feat) in enumerate(pairs, 1):
    plt.subplot(1, 3, i)
    sns.scatterplot(x=df[x_feat], y=df[y_feat], hue=df['species'])
    plt.title(f'Bivariate Analysis: {x_feat} vs {y_feat}')
plt.tight_layout()
plt.show()
```
